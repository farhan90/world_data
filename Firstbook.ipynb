##Relationship between Life Expectancy and Commercial Prime Lending Rate

#A map of the world with the location of their capitals

Since I had the data to pinpoint the exact location of the capital cities I decided to map it on the world map. I used the basemap
framework of python and its a useful tool for plotting maps.


```python
from mpl_toolkits.basemap import Basemap



def convert_to_float(x):
        dir_dict={'N':1,'S':-1,'E':1,'W':-1}
        s=str(x);
        l=s.split(' ')
        if(l[0]!='nan'):
            return (int(l[0])+(int(l[1])/60))*dir_dict[l[2]]


x=df[['Capital City','Latitude of Capital','Longitude of Capital']]

#x['Latitude of Capital']=x['Latitude of Capital'] != 'NA'

fig, ax = plt.subplots(figsize=(24,18))
map = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80,
            llcrnrlon=-180, urcrnrlon=180, resolution='l')
 
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='green')

map.drawmapboundary()
 
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

ax.set_title('World Map')
map.ax = ax


lats=x['Latitude of Capital'].map(convert_to_float).tolist()
lons=x['Longitude of Capital'].map(convert_to_float).tolist()

y,z = map(lons, lats)
map.plot(y, z, 'ro', markersize=6)
plt.show()
```

#OLS Model to predict GDP per capita

I decided to model the estimated GDP per capita on 2 variables which literacy rate and life expectancy. I think those 2 variables can give an indication how much a person should make in income, because if you are educated and live longer you will make more.

Before running the model I had to deal with missing values. In this case too I replaced missing values with the average. Though this is not the best technique I think it is sufficient in this case.

From the OLS model the coefficient for literacy rate is 173.1414 and the coefficient for Life Expectancy is 39.2692

We can see that most third world countries fall below their predicted GDP per capita, a reason for this is because bad political policies and unstability harms the economic output of that country. 


```python
import statsmodels.api as sm
x=df
clean_data('Literacy Rate (%)',x)
gdp_percap=x['GDP (PPP) in US $']/x['Population']

y=gdp_percap
z=x[['Literacy Rate (%)','Life Expectancy']]

res=sm.OLS(y,z)

res=res.fit()
res.summary()
```

```python

y_hat=res.predict()
x['Estimated GDP per capita']=y_hat
x['GDP Per Capita']=gdp_percap

x=x[['Country Name','Literacy Rate (%)','Life Expectancy','GDP Per Capita','Estimated GDP per capita']]
x
```

#Social Welfare Ranking

The variables I thought were important for measuring the social welfare of a country are as follows:
1) Life Expectancy - A happy society I believe tend to live longer
2) GDP per capita - To measure the income of an average individual
3) Health Expenditure - To measure the health care provided in that country
4) Literacy rate - To measure how educated a country is
5) Population density - To measure if the country is over populated or not

social welfare= 0.3*Life expectancy + 0.3*GDP Per Capita + 0.15*Health Expenditure +0.15* Literacy Rate-0.1*population density

The table below shows the rankings of the countries according to their social welfare score.

According to my social welfare function the best place to live would be Monaco and the worst place to live would be democratic republic of Congo.

I think this function does a decent job in predicting the social welfare of country since from the results it can be seen that small rich countries are a decent place to live.



```python
x=df

def score(var,weight):
    return var*weight;


x['GDP Per Capita']=gdp_percap
x['Population Density']=x['Population']/x['Area (sq km)']

clean_data('Literacy Rate (%)',x)
clean_data('Health Expenditure/GDP',x)

x['Social Welfare Score']=x['Life Expectancy'].apply((lambda x : score(x,0.3)))+ x['GDP Per Capita'].apply((lambda x:score(x,0.3)))+ x['Health Expenditure/GDP'].apply((lambda x : score(x,0.15)))+x['Literacy Rate (%)'].apply((lambda x:score(x,0.15)))-x['Population Density'].apply((lambda x:score(x,0.1)))

x[['Country Name','Social Welfare Score']].sort(columns='Social Welfare Score',ascending=False)
```

#Nearest Neighbor

I have used the same set variables to calculate the distance function as the social welfare function. My idea was countries with higher GDP, literacy rate and life expectancy,population and land area will be closer to each other. 

The table below show's the results.
One of the problem is computing this data is to know how to normalize data, since population data is so large it eclipses something like life expectancy or literacy rate.





```python
import math
x=df

def dv(x):
    return x/1000

def dv2(x):
    return x/100000

clean_data('Literacy Rate (%)',x)

x['GDP Per Capita']=gdp_percap
x['Distance Measure']=x['GDP Per Capita'].map(dv)+x['Literacy Rate (%)']+x['Life Expectancy']+x['Population'].map(dv2)+ x['Area (sq km)'].map(dv)
x=x[['Country Name','Distance Measure']]


x['Farthest Neighbor']=pd.Series()
x['Closest Neighbor']=pd.Series()


def calc_dist(x,y):
    return math.sqrt((x - y)**2 + (x - y)**2)


#row index 0 is the country name
#row index 1 is the distance measure
#row index 2 is max country
#row index 3 is min country

for i,row in x.iterrows():
    mndst=10000
    mxdst=-1
    mn_cntry=""
    mx_cntry=""
    for j,jrow in x.iterrows():
        if(jrow[0]==row[0]):
            continue
        else:
            dst=calc_dist(row[1],jrow[1])
            if(dst<mndst):
                mndst=dst
                x.ix[i,'Closest Neighbor']=jrow[0]
                
            if(dst>mxdst):
                mxdst=dst
                x.ix[i,'Farthest Neighbor']=jrow[0]
                                
x
```
